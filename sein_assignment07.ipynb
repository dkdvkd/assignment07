# import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

# given data

```python
num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

# plot the noisy data(x, y1)

```python
plt.plot(x, y2, 'k.')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Noisy data")
plt.show()
```

# plot clean data(x, y2)

```python
plt.plot(x, y1, 'r.')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Clean data")
plt.show()
```

# find approximating line(using lib)

```python
from sklearn.linear_model import LinearRegression
def fit_polynomial(x, y, degree):
    '''
    Fits a polynomial to the input sample.
    (x,y): input sample
    degree: polynomial degree
    '''
    model = LinearRegression()
    model.fit(np.vander(x, degree + 1), y)
    return model

def apply_polynomial(model, x):
    '''
    Evaluates a linear regression model in an input sample
    model: linear regression model
    x: input sample
    '''
    degree = model.coef_.size - 1
    y = model.predict(np.vander(x, degree + 1))
    return y

model = fit_polynomial(x, y2, 9)
print(model)
p_y = apply_polynomial(model, x)
print(p_y)

plt.plot(x, y2, 'k.', label='original data')
plt.plot(x, y1, 'r.', label='clean data')
plt.plot(x, p_y,'b.', label='fitted line')

plt.legend()
plt.show()

```

# plot the polynomial curves(fit the noisy data by the least square error)

```python
import numpy.polynomial.polynomial as poly

for i in range(0,10):
    c = poly.polyfit(x, y2, i)
    f = poly.polyval(x, c)
    print("p: ", i)
    plt.plot(x, y2, 'k.', label='original data')
    plt.plot(x, y1, 'r', label='clean data')
    plt.plot(x, f, 'b', label='fitted line')
    plt.legend
    plt.show()
```
